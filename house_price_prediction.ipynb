## 1.Handling missing values

#### final_df contains only 2919 and there are many features have a lot of missing values corresponding to dataframe size  , so we want to delete this.



```python
final_df.drop(['PoolQC','MiscFeature','Alley','Fence','MasVnrType','SalePrice','FireplaceQu'],axis=1,inplace=True)
```

```python
final_df.shape
```

### Handling missing values with categorical variablesÂ¶

```python
categorical_variable=[]
for col in final_df.columns:
    if final_df[col].dtype==object:
        categorical_variable.append(col)
```

```python
print(categorical_variable)
```

```python
## we impute the categorical varieable with mode
 
for col in categorical_variable:
    
    final_df.fillna({col:final_df[col].mode()},inplace=True)
```

```python
## we impute the categorical varieable with mode
 
for col in categorical_variable:
    
    final_df.fillna({col:final_df[col].mode()},inplace=True)
```
